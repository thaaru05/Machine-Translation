```python
# Install required libraries
!pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
!pip install transformers
!pip install git+https://github.com/openai/whisper.git
!pip install sentencepiece
!pip install gtts

# Import necessary libraries
import whisper
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
from gtts import gTTS
import warnings
import os

warnings.filterwarnings('ignore')

# Upload your audio file
from google.colab import files
uploaded = files.upload()

# Load Whisper model and transcribe audio
model = whisper.load_model("base").to("cuda")
result = model.transcribe("fireship_bigO.mp3")  # Use the uploaded file name
english_text = result["text"]

# Translate the text
src_lang = 'en_XX'
translate_lang = 'ta_IN'
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")

tokenizer.src_lang = src_lang
encoded_text = tokenizer(english_text, return_tensors="pt")
generated_tokens = model.generate(
    **encoded_text,
    forced_bos_token_id=tokenizer.lang_code_to_id[translate_lang]
)
translated = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)

# Print the translated text
print(translated[0])

# Convert the translated text to speech
tts = gTTS(translated[0], lang='ta')
tts.save("translated_audio.mp3")

# Optionally, play the audio in Colab
from IPython.display import Audio
Audio("translated_audio.mp3")
```
